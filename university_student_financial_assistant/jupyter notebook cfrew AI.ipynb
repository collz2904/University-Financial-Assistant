```python
import os
from crewai import Agent, Task, Crew, Process

# Set your LLM provider (using GPT-4o-mini as seen in your screenshot)
os.environ["OPENAI_API_KEY"] = "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"

## --- AGENTS ---

budget_forecaster = Agent(
    role='Budget Forecast Agent',
    goal='Analyze student monthly income/expenses and predict the next 3 months.',
    backstory='You are an expert financial planner specializing in student budgets. '
              'You excel at identifying spending patterns and forecasting future balances.',
    verbose=True,
    allow_delegation=False
)

investment_advisor = Agent(
    role='Savings & Investment Advisor',
    goal='Provide personalized savings and beginner investment advice based on budget forecasts.',
    backstory='A cautious yet insightful financial advisor who helps students grow their '
              'small savings into meaningful investments.',
    verbose=True,
    allow_delegation=False
)

summary_generator = Agent(
    role='Summary & Visualization Generator',
    goal='Synthesize all financial analysis into a clear, student-friendly report.',
    backstory='A data storyteller who turns complex numbers into simple, actionable insights.',
    verbose=True,
    allow_delegation=False
)

## --- TASKS ---

task1 = Task(
    description='Analyze the student\'s monthly income of $1,500 and expenses of $1,200. '
                'Provide a detailed 3-month forecast of their balance.',
    expected_output='A monthly breakdown of projected balances for the next 90 days.',
    agent=budget_forecaster
)

task2 = Task(
    description='Based on the 3-month forecast, suggest 3 specific savings strategies '
                'and 2 low-risk investment options suitable for a student.',
    expected_output='A list of savings strategies and investment recommendations.',
    agent=investment_advisor,
    context=[task1] # This ensures the advisor sees the budget data
)

task3 = Task(
    description='Create a final Student Financial Report. Include a summary of the '
                'forecast and the investment advice in a supportive tone.',
    expected_output='A clean, formatted markdown report ready for the student.',
    agent=summary_generator,
    context=[task1, task2]
)

## --- CREW ---

student_finance_crew = Crew(
    agents=[budget_forecaster, investment_advisor, summary_generator],
    tasks=[task1, task2, task3],
    process=Process.sequential, # Matches your visual flow
    verbose=True
)

# Execute the project
result = student_finance_crew.kickoff()

print("\n\n########################")
print("## FINAL REPORT ##")
print("########################\n")
print(result)
```
